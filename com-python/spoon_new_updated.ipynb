```python
import sys
import numpy as np
import skimage.color
import skimage.filters
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage import io
from scipy import stats
```

```python
image = skimage.io.imread("IMG_15.jpg")
%matplotlib inline
io.imshow(image)
```

```python
sigma = np.std(image)
print(sigma)
```

```python
blur = skimage.color.rgb2gray(image)
blur = skimage.filters.gaussian(blur, sigma=np.std(blur), multichannel=False)
```

```python
t = skimage.filters.threshold_otsu(blur)
mask = blur > t
```

```python
sel = np.zeros_like(image)
sel[mask] = image[mask]
io.imshow(sel)
```

```python
grayscale = rgb2gray(sel)

io.imshow(grayscale)
```

```python
rows,cols = grayscale.shape[0:2]
print(rows,cols)
```

```python
listy = []
for x in range (1,rows):
        for y in range (1,cols):
            pixel = grayscale[x,y]
            #print (pixel)
            if pixel > 0.0:
                listy.append(pixel)
#print(listy)
```

```python
count = len(listy); print(count)

mean = np.mean(listy); print(mean)

median = np.median(listy); print(median)

mode = stats.mode(listy); print(mode)
```

```python
plt.hist(listy, 100)
plt.show()
```

```python
print(np.max(listy))

print(np.min(listy))

print(np.sort(listy))
```
